## Step 1

Create a dictionary in which the keys are unique words and the values are the number of times that word appears in the play.

```python
import matplotlib.pyplot as plt
import math
infile = open("romeoandjuliet.txt","r")
wordList = []
# create a dictionary to store the unique words and the values of them
wordDict = {}
counter = 0
punc = [".", "?", "\'", "\"", ";", ",", "\\", "-", "!", "[", "]", ":", "(", ")"]
for line in infile:
    wordList = line.split()
    for word in wordList:
        if word[-1] in punc:    
            word = word[:-1]  
        word = word.lower()
        wordDict.setdefault(word, 0)
        wordDict[word] += 1
```

## Step 2

Plot a bar graph that shows the frequency with which each word appears. What are the ten most frequent words? What about the least frequent? Comment on what this might mean.

```python
# plot the bar graph of each word
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams['savefig.dpi'] = 300 
x = wordDict.keys()
y = wordDict.values()
plt.bar(range(len(x)), y)
plt.xticks(range(len(x)), tuple(x))
plt.xlabel("unique words")
plt.ylabel("word count")
plt.show()
```

```python
# find the distinct number of count of different words
distinctVal = list(set(wordDict.values()))
dsctSortedValue = [v for v in sorted(distinctVal)]
dsctSortedValue.reverse()

# create 2 lists
# sortKeyList for storing unique words and sortValList for storing values
sortKeyList = []
sortValList = []

for v in dsctSortedValue:
    for key in wordDict.keys():
        if wordDict[key] == v:
            sortKeyList.append(key)
            sortValList.append(v)

# most frequent 10 words 
print("most frequent 10 words are as follow:")
mostWord = []
mostWordCount = []
for i in range(10):
    print(sortKeyList[i], sortValList[i])
    mostWord.append(sortKeyList[i])
    
# plot the bar graph of each word
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams['savefig.dpi'] = 300 
plt.bar(mostWord, sortValList[:10])
plt.xlabel("unique words")
plt.ylabel("word count")
plt.show()
```

```python
# least frequent words
print("least frequent words are as follow:")
leastVal = sortValList[-1]
leastWord = []
for i in range(len(sortKeyList)):
    if sortValList[i] == leastVal:
        print(sortKeyList[i], sortValList[i])
        leastWord.append(sortKeyList[i])
```

## Step 4
sort the dictionary according to decreasing word count. Plot the word count vs. the rank of the word (i.e. the most frequent word will be rank=1, the second most frequent word will be rank=2, etc). Try plotting this on a log-log scale (i.e. $log[word count]$ vs. $log[rank]$). What does this look like? Can you try to interpret this?

```python
plt.subplot(211)
plt.scatter(range(1,len(sortValList)+1), sortValList)
plt.ylabel("word count")
plt.xlabel("rank")
plt.title("relationship between rank and word count")

plt.subplot(212)
logCount = [math.log(y) for y in sortValList]
logRank = [math.log(y) for y in range(1,len(sortValList)+1)]
plt.scatter(logRank, logCount)
#plt.loglog(range(1,len(sortValList)+1), sortValList)
plt.ylabel("log[word count]")
plt.xlabel("log[rank]")
plt.title("relationship between log[rank] and log[word count]")
plt.show()
```

The log-log plot looks like $\frac{1}{a + e^{x}}$. Thus, it means just a small part of the all unique words conquer most of the word count. Which means there are some words being used at most of time. Also, according to the 10 most frequent words, we can know that the words like "and", "the", "i" etc repeat frequently. In contrast, there are most of words which only being used once or twice.
